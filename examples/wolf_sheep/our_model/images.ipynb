```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
paths = ["./exp/experiment_w_10_s_10_seed_7123", "./exp/experiment_w_10_s_10_seed_1287", "./exp/experiment_w_10_s_10_seed_6372", "./exp/experiment_w_10_s_10_seed_2651", "./exp/experiment_w_10_s_10_seed_199"]

fig, axes = plt.subplots(3, 1, figsize=(10, 10))

for path in paths:
    df = pd.read_csv(path)
    sns.lineplot(df['Wolves'], ax=axes[0])
    sns.lineplot(df['Sheep'], ax=axes[1])
    sns.lineplot(df['Grass'], ax=axes[2])

axes[0].set_title('Wolves')
axes[0].set_ylabel('Count')
axes[1].set_title('Sheep')
axes[1].set_ylabel('Count')
axes[2].set_title('Grass')
axes[2].set_ylabel('Count')
for ax in axes:
    ax.set_xlabel('Generations')
axes[0].legend(title='Seed')
for ax in axes:
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
paths = ["./exp/experiment_w_10_s_10_seed_7123", "./exp/experiment_w_10_s_10_seed_1287", 
         "./exp/experiment_w_10_s_10_seed_6372", "./exp/experiment_w_10_s_10_seed_2651", 
         "./exp/experiment_w_10_s_10_seed_199"]

# Lista para armazenar os DataFrames
dfs = []

# Iterar sobre os caminhos e calcular a mdia e o desvio padro
for path in paths:
    df = pd.read_csv(path)
    dfs.append(df)

# Concatenar os DataFrames
concatenated_df = pd.concat(dfs)

# Calcular a mdia e o desvio padro para cada ponto de dados
mean_values = concatenated_df.groupby(concatenated_df.index).mean()
std_values = concatenated_df.groupby(concatenated_df.index).std()

# Criar uma figura e subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 10))

# Plotar a mdia e o desvio padro
sns.lineplot(x=mean_values.index, y=mean_values['Wolves'], ax=axes[0], label='Mean')
axes[0].fill_between(mean_values.index, mean_values['Wolves'] - std_values['Wolves'], mean_values['Wolves'] + std_values['Wolves'], alpha=0.3)
sns.lineplot(x=mean_values.index, y=mean_values['Sheep'], ax=axes[1], label='Mean')
axes[1].fill_between(mean_values.index, mean_values['Sheep'] - std_values['Sheep'], mean_values['Sheep'] + std_values['Sheep'], alpha=0.3)
sns.lineplot(x=mean_values.index, y=mean_values['Grass'], ax=axes[2], label='Mean')
axes[2].fill_between(mean_values.index, mean_values['Grass'] - std_values['Grass'], mean_values['Grass'] + std_values['Grass'], alpha=0.3)

# Adicionar ttulos e rtulos
axes[0].set_title('Wolves')
axes[0].set_ylabel('Count')
axes[1].set_title('Sheep')
axes[1].set_ylabel('Count')
axes[2].set_title('Grass')
axes[2].set_ylabel('Count')
for ax in axes:
    ax.set_xlabel('Time')

plt.tight_layout()
plt.show()
```
